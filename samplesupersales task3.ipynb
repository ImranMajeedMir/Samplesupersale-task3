```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
import warnings
warnings.filterwarnings('ignore')
# for interactiveness
from ipywidgets import interact
```

```python
# Read the dataset
d = pd.read_csv('Desktop/SampleSuperstore.csv')
# lets see the first 6 rows of the dataset
d.head(6)
```

```python
# Lets Check missing values
d.isnull().sum()
```

```python
# name of ship mode present our dataset
d['Ship Mode'].value_counts()
```

```python
# Check for duplicates
d.duplicated().sum()
```

```python
# Drop duplicates
d.drop_duplicates(inplace=True)
```

```python
# Check the shape of the dataframe
print("shape of dataset:",d.shape)
```

```python
# Lets calculate summary statistics
d.describe()
```

```python
# Now we will visualize the data using histograms
d.hist(bins=12, figsize=(10,15))
```

```python
# Lets analyze the data by category
d.groupby('Category')[['Sales', 'Profit']].sum().plot(kind='bar', figsize=(10,5))
```

```python
# Lets visualize the data using scatterplots
sns.scatterplot(x='Sales', y='Profit', data=d)
```

```python
# Now we will analyze the data by region
d.groupby('Region')[['Sales', 'Profit']].sum().plot(kind='bar', figsize=(10,5))
```

```python
#Lets analyse the correlation between each variable
plt.figure(figsize=(7,7),dpi = 100)
sns.heatmap(d.corr(),annot = True)
```

```python
# Lets visualize the data using boxplots
sns.boxplot(x='Category', y='Sales', data=d)
```

```python
# Analyze the data by sub-category
d.groupby('Sub-Category')[['Sales', 'Profit']].sum().plot(kind='bar', figsize=(10,5))
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(9,5),dpi = 200)
plt.title('sales vs Profit')
plt.plot(d['Sales'], label = 'Sales')
plt.plot(d['Profit'], label = 'Profit')
plt.legend()
```
